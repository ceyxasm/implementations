```python
#LAW OF LARGE NUMBERS

import random
import numpy as np
import matplotlib.pyplot as plt

random.seed(0)

def roll():
  return random.choice([1, 2, 3, 4, 5, 6])

def observation(i):
  count=0
  for j in range(i):
    if roll()==1:
      count+=1
  count=count/i
  prob.append(count)
  

for loop in range(6):
  prob=[]
  iter=[10, 100, 1000, 10000, 100000, 10**6]
  for i in iter:
    observation(i)
  plt.plot(prob)
plt.axhline(y = 1/6, color = 'r', linestyle = '-')
plt.show()

for loop in range(3):
  prob=[]
  iter=[100, 200, 300, 400, 500, 1000, 3000, 5000, 10**6]
  for i in iter:
    observation(i)
  plt.plot(prob)
plt.axhline(y = 1/6, color = 'black', linestyle = '-')
plt.show()
```

It is worth noting that the law of large numbers does not imply, as too many 
seem to think, that if deviations from expected behavior occur, these deviations 
are likely to be evened out by opposite deviations in the future. This 
misapplication of the law of large numbers is known as the **gambler’s fallacy**
>iTunes shuffle

>10^100000 will have substrings of million 0
In fact, if every subsequence of a large 
sequence of events appears to be random, it is highly likely that the sequence 
itself is not truly random.

The absolute difference from mean still deviates.

```python
#10000 trials of 100 flip each
count=[0,0]
abs_dif=[0]
ratio=[0]
random.seed(0)

def throw():
  for i in range(100):
    count[random.choice([0,1])]+=1
  abs_dif.append( abs(count[0]-count[1]))
  ratio.append(count[1]/count[0])

for i in range(10000):
  throw()
for i in range(1,len(abs_dif)):
  abs_dif[i]+=abs_dif[i-1]
plt.plot(abs_dif)
plt.show()

plt.plot(ratio)
plt.xscale('log')
plt.show()
  
```

```python
def std(lst):
  ar=np.array(lst)
  mean=np.mean(ar)
  sum=0
  for i in lst:
    sum+=(i-mean)**2
  return sum/len(lst)
```
