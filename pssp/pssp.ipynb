# **LAW OF LARGE NUMBERS**

In probability theory, the [law of large numbers](https://en.wikipedia.org/wiki/Law_of_large_numbers) (LLN) is a theorem that describes the result of performing the same experiment a large number of times. According to the law, the average of the results obtained from a large number of trials should be close to the expected value and will tend to become closer to the expected value as more trials are performed.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
random.seed(24091501)
```

```python


def roll(): #to simulate a rolling die
  return random.choice([1, 2, 3, 4, 5, 6])

def observation(i): #a die is rolled i times and observed mean of the number of 1s is calculated
  count=0
  for j in range(i):
    if roll()==1:
      count+=1
  count=count/i
  prob.append(count)
  

for loop in range(6): #so we make observation at 10, 100, .... throws
  prob=[]
  iter=[10, 100, 1000, 10000, 100000, 10**6]
  for i in iter:
    observation(i)
  plt.plot(prob)
plt.axhline(y = 1/6, color = 'r', linestyle = '-')
plt.show()

for loop in range(3): #once again but this time a closer look at hunder and thousand range
  prob=[]
  iter=[100, 200, 300, 400, 500, 1000, 3000, 5000, 10**6]
  for i in iter:
    observation(i)
  plt.plot(prob)
plt.axhline(y = 1/6, color = 'black', linestyle = '-')
plt.show()
```

As to be seen clearly, initially the observations were far from expected values but as the number of trials increased; observed values tended to expected values.

The beauty behind this??

The law of large numbers does not imply, as too many 
seem to think, that if deviations from expected behavior occur, these deviations 
are likely to be evened out by opposite deviations in the future.   This 
misapplication of the law of large numbers is known as the **[gamblers fallacy](https://www.gamblingnews.com/blog/how-does-the-law-of-large-numbers-apply-to-gambling-casinos/)**

All because somewhere, large is a relative concept and that if u were to flip a coin 10^(10^6) times, u r going to have lots of substrings of million 0s. Observing locally can make u jump to incorrect conclusions.

In fact, if every subsequence of a large
sequence of events appears to be random, it is highly likely that the sequence
itself is not truly random.

```python
#10000 trials of 100 flip each
count=[0,0]
abs_dif=[0]
ratio=[0]
random.seed(24091501)

def throw():
  for i in range(100):
    count[random.choice([0,1])]+=1
  abs_dif.append( abs(count[0]-count[1]))
  ratio.append(count[1]/count[0])

for i in range(10000):
  throw()
for i in range(1,len(abs_dif)):
  abs_dif[i]+=abs_dif[i-1]
plt.plot(abs_dif)
plt.title("absolute difference between the count of heads and tails")
plt.xlabel("number of trails")
plt.show()

plt.plot(ratio)
plt.xscale('log')
plt.title("ratio of count of heads to tails")
plt.show()
  
```

```python
def std(lst):
  ar=np.array(lst)
  mean=np.mean(ar)
  sum=0
  for i in lst:
    sum+=(i-mean)**2
  return sum/len(lst)
```

```python
def flip_ratio(num):
 heads = 0.0
 for i in range(num):
  if random.random() < 0.5:
    heads += 1
 return heads/num

def flipSim(num, trials):
 fracHeads = []
 for i in range(trials):
  fracHeads.append(flip_ratio(num))
 mean = sum(fracHeads)/len(fracHeads)
 sd = std(fracHeads)
 return (fracHeads, mean, sd)


def plot(num1, num2, trials):
 val1, mean1, sd1 = flipSim(num1, trials)
 plt.hist(val1, bins = 20)
 plt.title(str(trials)+ " trials of "+ str(num1)+ " flips each")
 plt.text(0.3, 10000, "mean="+str(round(mean1,5))+ "\n SD="+str(round(sd1,5)), fontsize = 12) 
 plt.show()
 val2, mean2, sd2 = flipSim(num2, trials)
 plt.hist(val2, bins = 20)
 plt.title(str(trials)+ " trials of "+ str(num2)+ " flips each")
 plt.text(0.44, 10000, "mean="+str(round(mean2,5))+ "\n SD="+str(round(sd2,5)), fontsize = 12) 
 plt.show()

random.seed(24091501)
plot(100,1000,100000)
```

And there we have our bell-curve.
Increasing the number of flips per trials makes the spread tighter.
